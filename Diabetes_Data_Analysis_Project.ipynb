```python
#1.What is the distribution of the 'Pregnancies' column?
plt.hist(df['Pregnancies'], bins=15, edgecolor='k')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Frequency')
plt.title('Distribution of Pregnancies')
plt.show()
```

```python
#2.How does the 'Glucose' level vary across different outcomes?
plt.figure(figsize=(8, 6))
df.boxplot(column='Glucose', by='Outcome')
plt.xlabel('Outcome')
plt.ylabel('Glucose Level')
plt.title('Glucose Level Variation by Outcome')
plt.suptitle('')  # Remove default title
plt.xticks([1, 2], ['No Diabetes (0)', 'Diabetes (1)'])  # Replace outcome values with labels
plt.show()
```

```python
#3.What is the average 'BloodPressure' of individuals with diabetes ('Outcome' = 1)?
average_blood_pressure_diabetes = df[df['Outcome'] == 1]['BloodPressure'].mean()

print("Average Blood Pressure for Individuals with Diabetes:", average_blood_pressure_diabetes)
```

```python
#4.Is there a correlation between 'SkinThickness' and 'BMI'?
correlation_skin_bmi = df['SkinThickness'].corr(df['BMI'])

print("Correlation between SkinThickness and BMI:", correlation_skin_bmi)
```

```python
#5.What is the median 'Insulin' level for individuals without diabetes ('Outcome' = 0)?
median_insulin_no_diabetes = df[df['Outcome'] == 0]['Insulin'].median()

print("Median Insulin Level for Individuals without Diabetes:", median_insulin_no_diabetes)
```

```python
#6.How does the 'DiabetesPedigreeFunction' differ between age groups?
# Define age groups
age_bins = [20, 30, 40, 50, 60, 70, 80]
age_labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']


# Add a new column with age groups
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)


# Calculate the mean DiabetesPedigreeFunction for each age group
age_group_mean_dpf = df.groupby('AgeGroup')['DiabetesPedigreeFunction'].mean()


# Plotting the results
plt.bar(age_group_mean_dpf.index, age_group_mean_dpf.values,color='green')
plt.xlabel('Age Group')
plt.ylabel('Mean DiabetesPedigreeFunction')
plt.title('Mean DiabetesPedigreeFunction by Age Group')
plt.show()
```

```python
#8.What percentage of individuals in the dataset have diabetes ('Outcome' = 1)?
percentage_diabetes = (df[df['Outcome'] == 1].shape[0] / df.shape[0]) * 100

print("Percentage of Individuals with Diabetes:", percentage_diabetes)
```
